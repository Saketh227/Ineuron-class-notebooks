### Use rename method to change column names that you want to
   you can either pass a 
   
> **dictionary** to `rename the particular columns` or 
          
>**function name** which takes only one argument
               `this allows you to apply the function on every column name`
    
   #### This applies same for the index labels also'''

```python
import pandas as pd
data = {"roll no": ["16191A0242","16191A0239","16191A0240","16191A0249","16191A0256"],
       "name":["P E Ganesh", "R SaiKiran","A Sravan","E S Vijay","V Vinay Kumar"]}
df= pd.DataFrame(data)

df
```

```python
'''Using dictionary'''
df.rename(columns={"roll no":"Roll_no","name":"Name"},inplace=True)
df
```

```python
'''Using Function'''
def column_name_modifier(col):
    col = col.lower()
    col = col.replace("_","--")
    return col

df.rename(columns=column_name_modifier) # it will not update original data
```

```python
df.rename(index={0:"col_0",1:"col_1"})
```

```python
def index_manager(ind):
    return "col_"+ str(ind)

df.rename(index=index_manager)
```
